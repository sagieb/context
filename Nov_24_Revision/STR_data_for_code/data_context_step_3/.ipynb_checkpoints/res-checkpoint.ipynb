```python
import pandas as pd
import numpy as np
import os
import re
from glob import glob
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
libs = ['Msn2A','Msn2B','Msn2C','Msn2D','Msn2E','Msn2F','Msn2G','Msn2H']
# libs = ['Cbf1A','Cbf1B']
strain = 'Msn2'
norm_to = 10**6
thresh = 7
for lib in libs:
# lib = 'Msn2E'
    all_adapts = pd.read_csv('./libs_info/R1_adapters.fasta', header=None,index_col=0)
    lib_adapt = all_adapts.filter(regex=lib,axis=0).reset_index().applymap(lambda x: x.replace('>'+lib+'_',''))
    lib_adapt.columns = ['seq_names']
    res_files_path = './seq_count/'
    samp_info = pd.read_csv('./libs_info/samp_info.csv')
    lib_df = pd.DataFrame(index=lib_adapt["seq_names"])
    res_files = os.listdir(res_files_path)
    samp_number = samp_info.query("Strain==@strain & Library==@lib").shape[0]
    samp_list = []
    for line in samp_info.query("Strain==@strain & Library==@lib").iterrows():
        curr_patt = line[1]['Pool name']+'_S[0-9]{1,2}'+'\.'+str(line[1]['Well barcode number'])+'\.'
        curr_file = [f for f in res_files if re.search(curr_patt, f)][0]
        curr_res = pd.read_csv(os.path.join(res_files_path,curr_file), delimiter=' ',header=None,index_col=0)
        curr_res.reset_index(inplace=True)
        curr_res.columns = ['Name','Reads']
        curr_res.index = curr_res["Name"].apply(lambda x: x.replace(lib+'_',''))
        ordered_res = curr_res.loc[curr_res.index.drop('no_adapter'),"Reads"].reindex(lib_df.index)
        curr_sample_name = str(line[1]["Trans repeat"])+'_'+str(line[1]["Time point"])+'_'+str(line[1]["Time point repeat"])
        samp_list.append(curr_sample_name)
        lib_df = pd.concat([lib_df,ordered_res],axis=1)
    lib_df.columns = samp_list
    sns.heatmap(lib_df.corr())
    plt.show()

    norm_log = (lib_df/lib_df.sum()*norm_to).apply(np.log2)
    passed_thr = norm_log.loc[norm_log.median(axis=1)>thresh,:]
    passed_thr = passed_thr.filter(regex='_180_').sub(passed_thr.filter(regex='_0_').mean(axis=1),axis=0)
    passed_thr.loc[:,'on/off'] = passed_thr.reset_index().seq_names.apply(lambda x: x.split('_')[-1]).values
    sns.scatterplot(data=passed_thr,x='1_180_1',y='1_180_2',hue='1_180_3',style='on/off')
    plt.show()
    on = passed_thr.loc[passed_thr['on/off']=='on'].drop('on/off',axis=1).mean(axis=1)
    on.index = on.reset_index().seq_names.apply(lambda x: '_'.join(x.split('_')[:-1])).values
    off = passed_thr.loc[passed_thr['on/off']=='off'].drop('on/off',axis=1).mean(axis=1)
    off.index = off.reset_index().seq_names.apply(lambda x: '_'.join(x.split('_')[:-1])).values
    data = pd.concat([on,off],axis=1)
    fig,ax = plt.subplots(1,1)
    sns.scatterplot(data=data,x=0,y=1,ax=ax)
    my_max = max(np.max(ax.get_ylim()),np.max(ax.get_xlim()))
    my_min = min(np.min(ax.get_ylim()),np.min(ax.get_xlim()))
    ax.set_ylim(my_min,my_max)
    ax.set_xlim(my_min,my_max)
    ax.axline([0,0],slope=1)
    plt.show()
```

```python

avgs = pd.DataFrame(passed_thr.drop('on/off',axis=1).mean(axis=1))
avgs.loc[:,'groups'] =  passed_thr.drop('on/off',axis=1).mean(axis=1).reset_index().seq_names.apply(lambda x: '_'.join(x.split('_')[:-1])).values
d = avgs.groupby('groups').agg(lambda x: list(x)).to_dict()[0]
my_df = pd.DataFrame({key:pd.Series(value) for key, value in d.items()}).T
fig,ax = plt.subplots(1,1)
sns.scatterplot(data=my_df,x=0,y=1,ax=ax)
ax.axline([0,0],slope=1)
```
